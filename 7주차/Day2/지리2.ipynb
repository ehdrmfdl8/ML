```python
import pandas as pd
import folium
import numpy as np
import json

Seoul_loc = [37.5662952, 126.9779451]
df = pd.read_csv('pop_n_CCTV.csv', encoding = 'euc-kr', index_col = 0)
df.tail()
```

```python
# 지도 만들기
map = folium.Map(location = Seoul_loc, zoom_start = 11)

# 경계 데이터 읽기
geo_str = json.load(open('Seoul_geo.json', encoding = 'utf-8'))

# 경계 표시(인구)
map.choropleth(geo_data = geo_str, data = df.인구, fill_color = 'PuRd', key_on = 'feature.id')

# 원 마커 표시(CCTV)
for n in df.index:
    folium.CircleMarker([df['위도'][n], df['경도'][n]], radius = df.CCTV[n] / 20, fill = True).add_to(map)

map
```
